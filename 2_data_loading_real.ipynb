```python
import pandas as pd

demo = pd.read_sas("data/DEMO_J.xpt")
bp = pd.read_sas("data/BPX_J.xpt")
kidney = pd.read_sas("data/ALB_CR_J.xpt")

print("DEMO shape:", demo.shape)
print("BP shape:", bp.shape)
print("KIDNEY shape:", kidney.shape)

```

```python
# Now actual Cleaning of Data. Going to select those column that are actually needed and useful.
```

```python
# Demographics (Age, Gender, BMI)
# SEQN = participant ID 
# RIDAGEYR = age
# RIAGENDR = gender
# Cleaning demographic data
demo_sel = demo[['SEQN', 'RIDAGEYR', 'RIAGENDR']]
demo_sel.head()
```

```python
#  BP cleaning
bp_sel = bp[['SEQN', 'BPXSY1', 'BPXDI1']]
bp_sel.head()

```

```python
# Kidney  data cleaning 
kidney_sel = kidney[['SEQN', 'URXUCR', 'URXUMA']]
kidney_sel.head()
```

```python
# mergeing all dataset
#merge demographics and BP data
df = pd.merge(demo_sel, bp_sel, on='SEQN', how='inner')
print(df.shape)
```

```python
# Merge kidney data
df = pd.merge(df, kidney_sel, on='SEQN', how='inner')
print(df.shape)

```

```python
# Final Preview
df.head()
```

```python
# check for missing values 
df.isnull().sum()
```

```python
# removing missing values 
df_clean = df.dropna()
print("Before cleaning:", df.shape)
print("After cleaning:", df_clean.shape)
```

```python
# Final cleaned preview
df_clean.head()
```

```python
def kidney_risk(row):
    if row['URXUMA'] > 30 or row['URXUCR'] > 150:
        return 'High'
    elif row['URXUCR'] > 100:
        return 'Medium'
    else:
        return 'Low'
```

```python
df_clean['kidney_risk'] = df_clean.apply(kidney_risk, axis=1)
```

```python
df_clean['kidney_risk'].value_counts()
```

```python
df_clean.head()
#preview of final labeled data
```

```python
# changing column name
df_final = df_clean.rename(columns={
    'RIDAGEYR': 'age',
    'RIAGENDR': 'gender',
    'BPXSY1': 'systolic_bp',
    'BPXDI1': 'diastolic_bp',
    'URXUCR': 'urine_creatinine',
    'URXUMA': 'urine_albumin'
})
```

```python
# selecting final columns after renaming them
df_final = df_final[[
    'age',
    'gender',
    'systolic_bp',
    'diastolic_bp',
    'urine_creatinine',
    'urine_albumin',
    'kidney_risk'
]]
```

```python
df_final.head()
```

```python
# saving data
# Raw clinical codes were transformed into human-readable features to improve interpretability and modeling.
df_final.to_csv("kidney_dataset.csv", index=False)
print("Final dataset saved as kidney_dataset.csv")
```

```python
# Training process is going to start
import pandas as pd

df = pd.read_csv("kidney_dataset.csv")

print(df.shape)
df.head()
```

```python
# Setting features and target
X = df[[
    'age',
    'gender',
    'systolic_bp',
    'diastolic_bp',
    'urine_creatinine',
    'urine_albumin'
]]

y = df['kidney_risk']
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

label_encoder.classes_
```

```python
# Training, Testing and spliting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
# stratify isliye ke har class ka balance rahe.
```

```python
# Model Train on Random Forest
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
# Model Evaluation
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))
```

```python
import joblib

joblib.dump(model, "kidney_model.pkl")
joblib.dump(label_encoder, "label_encoder.pkl")

print("Model and encoder saved successfully")
```
